```python
import warnings

import numpy as np
import pandas as pd


def generate_log_returns(data, lag):
    log_returns = pd.Series(np.nan, index=data.index)

    # Compute log returns based on the rules
    for t in range(len(data)):
        with warnings.catch_warnings():
            warnings.simplefilter('ignore')
            try:
                log_returns.iloc[t] = np.log(data.iloc[t + lag + 1] / data.iloc[t + 1])
            except Exception:
                log_returns.iloc[t] = np.nan
    return log_returns


def generate_targets(column_a: pd.Series, column_b: pd.Series, lag: int) -> pd.Series:
    a_returns = generate_log_returns(column_a, lag)
    b_returns = generate_log_returns(column_b, lag)
    return a_returns - b_returns
```

```python
generate_targets(train['LME_CA_Close'], train['US_Stock_CCJ_adj_close'], 2)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import STL

# Load train.csv (sample)
train = pd.read_csv("data/train.csv")

# Example: pick one instrument column (say 'LME_CU')
series = train[['date_id', 'LME_AH_Close']].dropna()
# series['date_id'] = pd.to_datetime(series['date_id'])
series = series.set_index('date_id')

# If it's prices, convert to log returns for stationarity
series['log_return'] = series['LME_AH_Close'].pct_change().apply(lambda x: np.log1p(x))

# Drop NA
y = series['log_return'].dropna()

# Apply STL decomposition
stl = STL(y, period=250)  # '5' if you expect weekly seasonality, try 20/21 for monthly
result = stl.fit()
```

```python
trend_component = result.trend
fig = trend_component.plot()
plt.show()
```

```python
season_component = result.seasonal
fig = season_component.plot()
plt.show()
```

```python
resid_component = result.resid
fig = resid_component.plot()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
plt.figure(figsize=(12,5))

# Autocorrelation (ACF)
plt.subplot(1,2,1)
plot_acf(series, lags=50, ax=plt.gca())
plt.title("Autocorrelation (ACF)")

# Partial Autocorrelation (PACF)
plt.subplot(1,2,2)
plot_pacf(series, lags=50, ax=plt.gca(), method="ywm")
plt.title("Partial Autocorrelation (PACF)")

plt.tight_layout()
plt.show()
```
